```python
import math

# таблица пользователей на 1 миллиард строк
TOTAL_RECORDS = 1000000000

# В бинарном дереве каждый узел имеет максимум 2 потомка.
BINARY_TREE_BRANCHING_FACTOR = 2

# В PostgreSQL размер страницы (page/блок) равен 8 КБ.
# Если один ключ (например, BIGINT) + указатели весят около 40 байт,
# то в один узел B-дерева поместится примерно 200 ключей.
BTREE_BRANCHING_FACTOR = 200


def calculate_tree_depths():
    print(f"Анализ I/O сложности для {TOTAL_RECORDS:,} записей...\n")

    # ========================================================
    # TODO 1.1: Вычислите глубину сбалансированного бинарного дерева.
    # Используйте формулу логарифма по основанию 2 от количества записей.
    # (math.log2 или math.log)
    # Округлите вверх до целого числа (math.ceil), так как мы не можем
    # сделать "половину" чтения с диска.
    # ========================================================
    binary_tree_depth = math.ceil(math.log(TOTAL_RECORDS, BINARY_TREE_BRANCHING_FACTOR))

    print("--- Сбалансированное бинарное дерево (AVL / Red-Black) ---")
    print(f"Глубина дерева (max I/O operations): {binary_tree_depth}\n")

    # ========================================================
    # TODO 1.2: Вычислите глубину B-Tree.
    # В B-дереве логарифм берется не по основанию 2,
    # а по основанию фактора ветвления (BTREE_BRANCHING_FACTOR).
    # Округлите вверх до целого числа (math.ceil).
    # ========================================================
    btree_depth = math.ceil(math.log(TOTAL_RECORDS, BTREE_BRANCHING_FACTOR))

    print("--- B-Tree (Классическая реляционная БД) ---")
    print(f"Глубина дерева (max I/O operations): {btree_depth}\n")

    # ========================================================
    # TODO 1.3: Посчитайте, во сколько раз B-Tree эффективнее.
    # ========================================================
    if btree_depth > 0:
        difference = binary_tree_depth / btree_depth
        print(f"Вывод: B-Tree требует в {difference:.1f} раз меньше прыжков по диску.")
        print("Именно поэтому классические БД используют B-деревья!")


if __name__ == "__main__":
    calculate_tree_depths()
```
